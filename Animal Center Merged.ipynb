```python
# Import Dependencies
import pandas as pd

# Create a reference the CSV file desired
intake_csv = "Resources/Austin_Animal_Center_Intakes.csv"
outcome_csv= "Resources/Austin_Animal_Center_Outcomes.csv"

# Read the CSV into a Pandas DataFrame
intake_df = pd.read_csv(intake_csv)
outcome_df = pd.read_csv(outcome_csv)

intake_df=intake_df.add_prefix('IN_')
outcome_df=outcome_df.add_prefix("OUT_")

intake_df=intake_df.rename(index=str,columns={"IN_Animal ID": "Animal ID"})
outcome_df=outcome_df.rename(index=str,columns={"OUT_Animal ID": "Animal ID"})


animal_data_merged = pd.merge(intake_df, outcome_df, on="Animal ID", how="outer")
animal_data_merged
```
